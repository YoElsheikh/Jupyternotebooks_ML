```python
import numpy as np
import pandas as pd

def perceptron(x, w, b):

    #single perceptron
    linear_output = (np.dot(x, w) + b) #this will return a one dimentional array, the summation number
if linear_output > 0.0:b           #not necesssarily, check note below
    activation = 1
else:
    activation = 0 
        
return activation

def prediction(x, w, b):
    #the prediction function computes the prediction for each given example
    y_hat = np.array([])
    for x_i in x.values:
        y_hat = np.append(y_hat, perceptron(x_i, w, b))
    return y_hat 
```

## Training


```python
def perceptron_train(x, y, learning_rate = 0.01, epochs = 50):
    #using the training data (x, y) train the model
    n_features = x.shape[1] #no of input features 
    w = np.zeros(n_features) #initialize the weights vector with 0s
    b = 0
    for e in range (epochs): #finds optimal weights and biases for each element of the dataset, i.e. a weights vector
        for x_i, y_i in zip(x, y): #performs the calculation on all of the vector values and returns a weights vector and a bias???????
            y_hat = perceptron(x_i, w, b)
            w = w + learning_rate * (y_i - y_hat) * x_i #w + learning_rate*derivative of the error wrt the weights, y_hat/y_i is a singlevalue 0 or 1,the output
            b = b + learning_rate * (y_i - y_hat)
        return(w, b)
    #i find w and b that reduces the error
    #the perceptron returns one equatoion which is the equation of a line, i think
    #we then use this equation of a line for binary classification
        
       
```

```python
#loading the iris dataset

from sklearn import datasets 
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['type'] = iris.target #here these are the 3 outputs assigned values from 0 to 2

#df.sample(5) #view 3 random samples
df.tail(20)  #view the last 20 samples in contrast to df.head()
#df.head()


```

```python
print('Output classes before removal:', df['type'].unique())

df = df[df['type'] != 2].reset_index(drop=True)  #dropped one classification class because our perceptron's output is binary

print('Output classes before removal:', df['type'].unique())

len(df)
```

```python
#split the  dataset into training and testing
train, test = train_test_split(df, test_size = 0.2)

x_train = train.drop('type', axis = 1) #input the features - the type column (axis = 1 references a column)
y_train = train['type'] #what i want to predict

x_test = test.drop('type', axis = 1)
y_test = test['type']

print(len(x_train), len(x_test))



```

```python
#training the perceptron, returning the weights and biases, basically
alpha = 0.1
(w, b) = perceptron_train(x_train.values, y_train.values, alpha, epochs = 20)
```

```python
#predicting the type
prediction_df = x_test.copy()
y_hat = prediction(prediction_df, w, b) #will retturn a vector of the activations i.e. zeros and ones
prediction_df['type'] = y_hat
x_test['type'] = y_test
```
